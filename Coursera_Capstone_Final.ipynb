## Introduction 

The audience of the project is the people who never been to US before and want to take a cultural trip in the main cities of US. I select 2 main categories to reprensent the culture of one city, museums and colleges. Museums have great exhibits that record history and colleges are one of the main organizatios to cultivate talents and shape the future. The aim of the project is to compare the number of museums and colleges in Boston, MA, New York City, NY, Chicago, IL and Los Angeles, CA.

## Data

To finish the project, I use Foursquare API to get the information of museums and colleges in these 4 cities. The center/mean coordinates of 4 cities are provided by Foursquare.
