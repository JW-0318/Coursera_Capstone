```python
url_ny = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll=40.7127281, -74.0060152&radius=3000&limit=100&categoryId=4bf58dd8d48988d181941735'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION)
```

```python
url_ny
```

```python
results_ny = requests.get(url_ny).json()
results_ny
```

```python
venues_ny = results_ny['response']['groups'][0]['items']
    
nearby_venues_ny = json_normalize(venues_ny) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.location.lat', 'venue.location.lng']
nearby_venues_ny =nearby_venues_ny.loc[:, filtered_columns]

# clean columns
nearby_venues_ny.columns = [col.split(".")[-1] for col in nearby_venues_ny.columns]

nearby_venues_ny.head()
```

```python
print('In New York City, {} venues were returned by Foursquare.'.format(nearby_venues_ny.shape[0]))
```

```python
# create map
map_ny = folium.Map(location=[40.7127281, -74.0060152], zoom_start=13)

# add markers to map
for lat, lng, name in zip(nearby_venues_ny['lat'], nearby_venues_ny['lng'], nearby_venues_ny['name']):
    label = '{}'.format(name)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_ny)  
    
map_ny
```
